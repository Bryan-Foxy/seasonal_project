# Visualisation of images

## 1.Setup

```python
# Libraries
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## 2.Image Understanding

```python
# Figure
image = cv2.imread("images/bird 4k.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
h,w, c = image.shape
cv2.putText(image, f"height: {h}px, width: {w}px, channel: {c}", (100,300), cv2.FONT_HERSHEY_SIMPLEX, 5, (0,255,0), 2)

plt.title('Bird 4k')
plt.axis('off')
plt.imshow(image)
```
